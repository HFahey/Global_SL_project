```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import glob
%matplotlib inline
```

```python
files = glob.glob('./Data/*.csv')
df_list = []
for file_name in files:
    df = pd.read_csv(file_name, sep=',')
    df_list.append(df)
#df_list[1]
```

```python
cmap = plt.get_cmap('rainbow')
norm = mcolors.Normalize(vmin=0, vmax=(len(files))-1)

timestep=np.arange(0,len(files),1)
def get_colour(timestep):
    return cmap(norm(timestep))
c=cmap(norm(timestep))

```

```python
files=os.listdir('./Data')
for i in np.arange(0,len(df_list),1):
    plt.plot(df_list[i]['Age (Ma)'], df_list[i]['Depth (mbsf)']*-1, label=files[i], c=c[i])
plt.legend(loc='best')
plt.xlabel('Age (Ma)')
plt.ylabel('Depth (m)')
```

```python
fs=8
ms=2
fig, axs = plt.subplots(len(files),1, figsize=(7.5/2.5, 23/2.5))
    
for i in np.arange(0,len(files),1):
    ax = axs[i]
    ax.plot(df_list[i]['Age (Ma)'], df_list[i]['Depth (mbsf)']*-1, '-o', label=files[i], c='r', markersize=ms)
    ax.set_title(str(files[i]), pad=-10, fontsize=fs)
    ax.tick_params(axis='both', which='both', direction='in', labeltop=True, labelbottom=False, pad=0.1, top=True, right=True)
    ax.set_xlim(0,140)
    ax.set_ylim(-1300,200)
    ax.set_ylabel('Depth (m)')

axs[len(files)-1]
ax.set_xlabel('Age (Ma)')
```
